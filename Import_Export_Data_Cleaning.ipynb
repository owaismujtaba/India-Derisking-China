```python
import pandas as pd
import seaborn as sns 
import glob
```

```python
import_files = glob.glob("data/imports/country_wise/*.xlsx")
import_files = sorted(import_files)
import_files
```

```python
row = []
columns = ['sn', 'country', 'vol_p', 'share_p', 'vol_c', 'share_c', 'growth']
for file in import_files:
    print(file)
    df = pd.read_excel(file, header=1)
    df = df.iloc[:-1]
    year = int(df.columns[4].split('-')[1])
    df.columns = columns
    for index in range(df.shape[0]):
        row.append([df['country'][index], year, df['vol_c'][index],df['share_c'][index], df['growth'][index]])
    if year == 2019:
        for index in range(df.shape[0]):
            row.append([df['country'][index], year-1, df['vol_p'][index],df['share_p'][index], 'NA'])
```

```python
columns = ['Country', 'Year', 'Value(Imports)', 'Share', 'Growth']
import_data_all = pd.DataFrame(row, columns=columns)
import_data_all
```

```python
import_data_all.to_csv('clean_data/CountryWiseImports.csv')
```

```python
import os
import glob
import pandas as pd

base_folder = "data/imports/CommodityWiseAll"
columns = ['s.no', 'country', 'vol_p',  'vol_c', 'growth']
final_columns = ['Country','HSCode', 'Year', 'Value(Imports)', 'Growth']

# Loop through all subfolders
for subfolder in sorted(os.listdir(base_folder)):
    folder_path = os.path.join(base_folder, subfolder)
    if os.path.isdir(folder_path):
        print(f"Processing folder: {subfolder}")
        row = []  # reset for each folder
        excel_files = glob.glob(os.path.join(folder_path, "*.xlsx"))
        for file in sorted(excel_files):
            print(f"  Reading file: {os.path.basename(file)}")
            df = pd.read_excel(file, header=2)
            df = df.iloc[:-1, :5]  # keep relevant rows/columns
            year = int(df.columns[3].split('-')[0])
            df.columns = columns

            # Append data
            for index in range(df.shape[0]):
                row.append([df['country'][index], subfolder, year, df['vol_c'][index], df['growth'][index]])

            # Special case for 2019
            if year == 2019:
                for index in range(df.shape[0]):
                    row.append([df['country'][index], subfolder, year-1, df['vol_p'][index], 'NA'])
            
        # Save folder-specific CSV
        commodity_df = pd.DataFrame(row, columns=final_columns)
        output_path = f"clean_data/CountryWiseImports_{subfolder}.csv"
        commodity_df.to_csv(output_path, index=False)
    
        
```
